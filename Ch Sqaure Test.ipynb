```python
from scipy.stats import chi2_contingency
```

```python
chi2_contingency(pd.crosstab(df.Age,df.Drink))
```

```python
from scipy import stats
chi2_stat,p_val,dof,ex=stats.chi2_contingency(pd.crosstab(df.Age,df.Drink))
print("===chi Chquare Test=====")
print(chi2_stat)
print('\n')

print("===P Value===")
print(p_val)
print('\n')

print("===Degree of Freedom===")
print(dof)
print('\n')

print("===Contingency Table===")
print(ex)
print('\n')
```
